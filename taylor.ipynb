```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def taylor_sin(x, terms=10):
    # Approximating sin(x) using Taylor series
    approximation = 0
    for n in range(terms):
        term = (-1) ** n * (x ** (2 * n + 1)) / math.factorial(2 * n + 1)
        approximation += term
    return approximation
```

```python
# Create a range of x values from -2π to 2π
x_values = np.linspace(-2 * math.pi, 2 * math.pi, 400)
# Compute the sine values using numpy for comparison
true_sin_values = np.sin(x_values)
```

```python
# Compute Taylor series approximations for different number of terms
taylor_approximations = {terms: [taylor_sin(x, terms) for x in x_values] for terms in [1, 3, 5, 7, 10]}
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(x_values, true_sin_values, label='sin(x)', color='black', linewidth=2)
for terms, approx_values in taylor_approximations.items():
    plt.plot(x_values, approx_values, label=f'Taylor series (terms={terms})')
```

```python
plt.title('Taylor Series Approximation of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
